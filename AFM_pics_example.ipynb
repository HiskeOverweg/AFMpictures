```python
%matplotlib inline
import os
from dataClass import data
import AFM
import matplotlib.pyplot as plt
import numpy as np
```

```python
fileName = 'splitgates.006'

# read measured channels into a dictionary
# keys are strings with channel names
# values are numpy arrays with data, with an extra attribute (.u)
# for the unit of the data
# extent and extentUnit contain the size of the scan frame
channels, extent, extentUnit = AFM.createChannelDict(fileName)
```

```python
def axisBar(ax, extent):
    x1 = 0
    x2 = 1
    y = -0.04
    dy = 0.01
    plt.plot([x1, x2], [y, y], 'k', clip_on=False, transform=ax.transAxes)
    plt.plot([x1, x1], [y+dy, y-dy], 'k', clip_on=False,
             transform=ax.transAxes)
    plt.plot([x2, x2], [y+dy, y-dy], 'k', clip_on=False,
             transform=ax.transAxes)
    plt.text(0.5, y-dy, '%.1f um' % (extent*1E-3),
             horizontalalignment='center', verticalalignment='top',
             transform=ax.transAxes)


for k in channels.keys():
    fig = plt.figure()
    fig.set_size_inches(6, 5)
    ax = fig.add_subplot(111)
    plt.pcolormesh(channels[k], cmap=plt.cm.afmhot, rasterized=True)
    cbar = plt.colorbar()
    cbar.set_label(k.split()[0]+' (%s)' % channels[k].u)
    plt.xticks([])
    plt.yticks([])
    axisBar(ax, extent)
    plt.title(k)
```
